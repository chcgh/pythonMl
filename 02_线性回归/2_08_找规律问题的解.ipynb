```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=np.array([[1,1,1,1,1],
           [1,2,4,8,16],
           [1,3,9,27,81],
           [1,4,16,64,256],
           [1,5,25,125,625]])


x=np.array([1,2,3,4,5])
y=np.array([2,8,116,47,68])
```

```python
result=np.matmul(np.linalg.inv(A),y.T)
```

```python
print(result.T)
```

```python
[a,b,c,d,e]=result.T
```

```python
def fun(x):
    return a+b*x+c*np.power(x,2)+d*np.power(x,3)+e*np.power(x,4)
```

```python
print(fun(x))
```

```python
x0=np.linspace(1,5,100)
```

```python
y0=fun(x0)
```

```python

plt.plot(x, y)
plt.plot(x0, y0)
plt.xlabel("x")
plt.ylabel("y")
  
plt.show()
```
